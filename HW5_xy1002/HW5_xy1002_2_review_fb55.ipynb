````python
def residuals(p0, x1, x2, error = None):
    #you can add errors if you want 
    #you want to return the sum of the residuals 
    #(difference b/w data and model **2)
    for k in x2:
        residual = (line(x1, p0) - x2)**2
    return sum(residual)
````
